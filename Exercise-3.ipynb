### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame that should look like the one below.

```python
df = pd.DataFrame([['Jason', 'Miller', 42, 0, 4, 25], 
                   ['Molly', 'Jacobsen', 52, 1, 24, 94], 
                   ['Tina', 'Ali', 36, 1, 31, 57], 
                   ['Jake', 'Milner', 24, 0, 2, 62], 
                   ['Amy', 'Cooze', 73, 1, 3, 70]
                   ], 
                  columns=['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(x=df['postTestScore'], y=df['preTestScore'], s=df['age'])
plt.xlabel('Post test score')
plt.ylabel('Pre test Score')
plt.title("Pre Test score vs Post test score")
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter(x=df['postTestScore'], y=df['preTestScore'], s=df['postTestScore'] * 4.5, c=df['female'])
plt.xlabel('Post test score')
plt.ylabel('Pre test Score')
plt.title("Pre Test score vs Post test score 2")
```
